```python
import pandas as pd
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt

X,Y = make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,
                         n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=10)

plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=Y,cmap='winter')
```

```python
def step(X):
    return np.where(X >= 0, 1, 0)

def perceptron(X,Y):

    X=np.insert(X,0,1,axis=1)  #here 0,1 means at 0th index value will be 1 here it will be bias and 2 columns of x each will act as weight 
    weights = np.ones(X.shape[1]) #cosidering weight as 1 [1,1,1]
    lr = 0.1

    for i in range(1000): 
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j], weights)) #step function predicted op will be 0 or 1
        weights = weights + lr * (Y[j]-y_hat) *X[j]  #new weights

    return weights[0],weights[1:] # return bias and w1,w2
```

```python
bias, coef = perceptron(X,Y);

slope_m = -(coef[0]/coef[1])
intercept_b = -(bias/coef[1]) 

x_input = np.linspace(-3,3,100)
y_input = slope_m * x_input + intercept_b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input, y_input, color='red', linewidth=2)
plt.scatter(X[:,0],X[:,1],c=Y,cmap='winter')
plt.ylim(-3,2)
```
